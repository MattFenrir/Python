```python
# Faça um jogo da sorte que guarde um número (0 - 99) e
# peça para o usuário tentar acertar o número. Exiba se errou para mais ou menos
#1 Fácil -> 0 25
#2 Médio -> 0 -> 100
#3 Difícil -> 0 -> 100 -> Cara -> Não dou a dica
#4 ou > 4 -> Sair

from random import randint

continuar = True
segredo = randint(0, 100)
chances = 0
acertou = False

while(continuar):
  menu = input('**** JOGO DA SORTE **** \n 1 - Fácil \n 2 - Médio \n 3 - Difícil \n 4 - Sair \n R: ')
  if(menu.isnumeric()):
    menu = int(menu)
    if(menu > 0 and menu < 4):
      print('**** INÍCIO DA PARTIDA ****')
      acertou = False
      chances = 0
      if(menu == 1):
        print('**** FÁCIL ****')
        segredo = randint(0, 25)
        while(not acertou): # == False
          chute = int(input('Digite um número entre 0 e 25: ')) # Pega valor
          chances += 1 # Tive mais uma tentativa
          if(chute == segredo):
            print('**** FIM DE PARTIDA ****')
            print('Você acertou:', segredo)
            print('Com {} tentativas:' .format(chances))
            acertou = True
          elif(chute > segredo):
            print('Menos')
          else:
            print('Mais')
      elif(menu == 2):
        print('**** MÉDIO ****')
        segredo = randint(0, 100)
        while(not acertou): # == False
          chute = int(input('Digite um número entre 0 e 100: ')) # Pega valor
          chances += 1 # Tive mais uma tentativa
          if(chute == segredo):
            print('**** FIM DE PARTIDA ****')
            print('Você acertou:', segredo)
            print('Com {} tentativas:' .format(chances))
            acertou = True
          elif(chute > segredo):
            print('Menos')
          else:
            print('Mais')
      else:
        print('**** DIFÍCIL ****')
        segredo = randint(0, 100)
        while(not acertou): # == False
          chute = int(input('Digite um número entre 0 e 100: ')) # Pega valor
          chances += 1 # Tive mais uma tentativa
          if(chute == segredo):
            print('**** FIM DE PARTIDA ****')
            print('Você acertou:', segredo)
            print('Com {} tentativas:' .format(chances))
            acertou = True
          else:
            sera = randint(0, 1) #Cara/Cora Par/Impar Sim/Não 50% de chance
            if(sera == 0): # 0 -> dá dica
              if(chute > segredo):
                print('Menos')
              else:
                print('Mais')
            else: # Se não for 0 não dou dica
              print('NÃO POSSO TE AJUDAR!')
    else:
      print('FIM DE JOGO')
      continuar = False
  else:
    print('Digite um valor VÁLIDO!!')
```
